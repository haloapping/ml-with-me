# Dataset

```python
import pandas as pd
import seaborn as sns

status_train_df = pd.DataFrame(
                        data={
                            "nama": ["Elisabeth", "Ical", "Davin", "Andre", "Putri", "Apping", "Tabi", "Laily", "Rahel", "Yar"],
                            "height": [160, 159, 155, 156, 160, 165, 169, 170, 164, 166],
                            "weight": [65, 65, 55, 59, 60, 55, 57, 59, 60, 55],
                            "status": ["pacaran", "pacaran", "pacaran", "pacaran", "pacaran",
                                       "jomblo", "jomblo", "jomblo", "jomblo", "jomblo"]
                        },
                    )

status_test_df = pd.DataFrame(
                      data={
                          "nama": ["Gio", "Henry", "Cantika"],
                          "height": [164, 163, 161],
                          "weight": [61, 60, 62],
                          "status": ["jomblo", "jomblo", "pacaran"]
                      },
                  )

ax = sns.scatterplot(
    data=status_train_df,
    x="weight",
    y="height",
    hue="status")

ax.set(
    xlabel="Weight",
    ylabel="Height",
    title="Sebaran Status berdasarkan Tinggi dan Berat Badan (Train Dataset)"
);
```

```python
import matplotlib.pyplot as plt

plt.scatter(data=status_train_df, x="weight", y="height", label="train dataset")
plt.scatter(data=status_test_df, x="weight", y="height", label="test dataset")
plt.title("Sebaran Status berdasarkan Tinggi dan Berat Badan (Train and Test Dataset)")
plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.legend()
plt.show()
```

# Classification Model

## Transform target value

```python
from sklearn.preprocessing import LabelBinarizer

label_binarizer_status = LabelBinarizer()

y_train_lb = label_binarizer_status.fit_transform(status_train_df["status"].values)
y_train_lb = y_train_lb.flatten()

y_test_lb = label_binarizer_status.fit_transform(status_test_df["status"].values)
y_test_lb = y_test_lb.flatten()
```

## Split dataset

```python
X_train = status_train_df[["height", "weight"]].values
y_train = y_train_lb

X_test = status_test_df[["height", "weight"]].values
y_test = y_test_lb

X_train, y_train
```

```python
X_test, y_test
```

## Train

```python
from sklearn.svm import SVC

svc_model = SVC(random_state=42)
svc_model.fit(X_train, y_train)
```

## Evaluate

```python
from sklearn.metrics import classification_report

y_pred = svc_model.predict(X_test)
print(classification_report(y_pred, y_test))
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

params = {
    "kernel": ["rbf", "poly", "sigmoid"],
    "C": [0.5, 1, 10, 100],
    "gamma": ["scale", 1, 0.1, 0.01, 0.001],
}

grid_search = GridSearchCV(
    estimator=SVC(random_state=42),
    param_grid=params,
    n_jobs=4,
    verbose=1,
    scoring="accuracy"
)

grid_search.fit(X_train, y_train)
```

```python
print(f"Best index     : {grid_search.best_index_}")
print(f"Best estimator : {grid_search.best_estimator_}")
print(f"Best params    : {grid_search.best_params_}")
print(f"Best score     : {grid_search.best_score_}")
```

```python
cv_result = pd.DataFrame(grid_search.cv_results_).sort_values(by="rank_test_score")
cv_result
```

```python
from sklearn.svm import SVC

svc_model = SVC(C=0.5, gamma='scale', kernel='rbf', random_state=42)
svc_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = svc_model.predict(X_test)
print(classification_report(y_pred, y_test))
```
